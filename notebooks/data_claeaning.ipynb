```python
import pandas as pd
```

```python
# read csv file 
food = pd.read_csv("../data/raw/global_food_prices.csv")
inflation = pd.read_csv("../data/raw/inflation.csv", skiprows=4) # That 4 rows is just uptodate informations.
```

```python
# check the first 5 rows.
food.head()
```

```python
inflation.head()
```

```python
# check food datatype
food.info()
```

```python
# select only the row that we want eg. country,product,price,month,year,unit
food_clean = food[['adm0_name', 'cm_name', 'mp_price', 'mp_month', 'mp_year', 'um_name']]
```

```python
# rename to common name
food_clean = food_clean.rename(columns = {
    'adm0_name' : 'country',
    'cm_name' : 'product',
    'mp_price' : 'price', 
    'mp_month' : 'month', 
    'mp_year' : 'year', 
    'um_name' : 'unit'
})
```

```python
# clean food data
food_clean
```

```python
# check infaltion datatype
inflation.info()
```

```python
# drop the column that we don't need
inflation_clean = inflation.drop(columns = ['Country Code','Indicator Name','Indicator Code','Unnamed: 70'])
```

```python
inflation_clean
```

```python
# change from wide format to long format (unpivot) column -> row ; 
# (pivot) row -> column
inflation_clean = inflation_clean.melt(
    id_vars=['Country Name'], # keep this
    var_name='year', # new column name
    value_name='inflation_rate' # new value name
)
```

```python
# rename columns country
inflation_clean = inflation_clean.rename(columns = {
    'Country Name': 'country'
})
```

```python
inflation_clean
```

```python
# change year to int type
inflation_clean['year'] = inflation_clean['year'].astype(int)
```

```python
# drop all na in inflation_rate
inflation_clean = inflation_clean.dropna(subset=['inflation_rate'])
```

```python
inflation_clean
```

```python
# merge using inner join
food_inflation = pd.merge(
    food_clean, # Always!!! use the one with most data
    inflation_clean,
    on=['country', 'year'],
    how='inner'
)
```

```python
food_inflation
```

```python
# check how many country are there
food_inflation['country'].unique()
```

```python
# see what product we have
food_inflation['product'].unique()
```

```python
# select only Thailand cause we will focusing only Thailand
thai_food_inflation = food_inflation[food_inflation['country'] == 'Thailand']
```

```python
thai_food_inflation
```

```python
thai_food_inflation['product'].unique()
```

```python
# save as a csv file
thai_food_inflation.to_csv("../data/cleaned_data/thai_food_inflation.csv", index=False)
```
